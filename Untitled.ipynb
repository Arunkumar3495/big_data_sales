```python
def find_outlier(col_name):
    Q1 = df[col_name].quantile(0.25)
    Q3 = df[col_name].quantile(0.75)
    IQR = Q3 - Q1
    
    outliers = df[(df[col_name]<Q1-1.5 * IQR) | (df[col_name] > Q3 +1.5*IQR)]
    
    print('Total coount of outliers are',outliers.shape[0])
    
```

```python
# visualize the data using box plot
sns.boxplot(x=df['Item_Weight'])
plt.show()

find_outlier('Item_Weight')
```

```python
sns.boxplot(x=df['Item_Visibility'])
plt.show()

find_outlier('Item_Visibility')
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Item_Identifier'] = encoder.fit_transform(df['Item_Identifier'])
df['Outlet_Identifier'] = encoder.fit_transform(df['Outlet_Identifier'])
df['Item_Type'] = encoder.fit_transform(df['Item_Type'])
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

Outlet_Size_encoded = onehotencoder.fit_transform(df['Outlet_Size'].values.reshape(-1,1)).toarray()

#get the list of encoding columns name

encoded_colums_names = onehotencoder.get_feature_names_out(['Outlet_Size'])

# create new data frame
train_cat_df = pd.DataFrame(Outlet_Size_encoded,columns = encoded_colums_names)

train_cat_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

Outlet_Size_encoded = onehotencoder.fit_transform(df['Outlet_Location_Type'].values.reshape(-1,1)).toarray()

#get the list of encoding columns name

encoded_colums_names = onehotencoder.get_feature_names_out(['Outlet_Location_Type'])

# create new data frame
train_cat_df = pd.DataFrame(Outlet_Size_encoded,columns = encoded_colums_names)

# train_cat_df.head()

df = pd.concat([df,train_cat_df],axis=1)

#droping the Outlet_location_Type

df = df.drop(['Outlet_Location_Type'],axis=1)
```
